```python
sharks = pd.read_csv('/content/wrangling/data/sharks.csv', low_memory=False)
sharks.head()
```

```python
type = sharks['Type']
type.value_counts()
#change unknown variables to NaN instead of values
type = type.replace(['Invalid', 'Questionable', 'Unconfirmed', 'Unverified', 'Under investigation' ], np.nan)
type.value_counts()

#group all boating incidents together
type = type.replace(['Sea Disaster', 'Watercraft', 'Boat', 'Boatomg', 'Boating'], 'Watercraft')
type.value_counts()

df['Type'] = type
df['Type'].value_counts() #this puts the revised type variable back into the dataset
```

```python
#Dummy variable: For the pretrial data covered in the lecture, clean the WhetherDefendantWasReleasedPretrial variable as well as you can, and, in particular, replace missing values with np.nan.

url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
trial = pd.read_csv(url,low_memory=False) # Pandas downloads and loads the .csv file for you
```

```python
release = trial['WhetherDefendantWasReleasedPretrial']
print(release.value_counts())
#the 9 is unclear in the cookbook
release = release.replace(9,np.nan) #repalce 9 with nan
print(release.value_counts())

#check that there are the correct number of missing values, 31
print('Missing values: ',sum(release.isnull()))

 #put the revised release variable back into dataset
trial['WhetherDefendantWasReleasedPretrial'] = release
```

```python
#Missing values, not at random: For the pretrial data covered in the lecture, clean the ImposedSentenceAllChargeInContactEvent variable as well as you can, and explain the choices you make. (Hint: Look at the SentenceTypeAllChargesAtConvictionInContactEvent variable.)

sentence = trial['ImposedSentenceAllChargeInContactEvent']
charge = trial['SentenceTypeAllChargesAtConvictionInContactEvent']
#print(charge.unique(), '\n')
#print(sentence.unique(), '\n') there are some missing values, some have '.'
sentence = pd.to_numeric(sentence,errors='coerce') #coerce to numeric value
sentence_NaN = sentence.isnull() #creates a dummy for missing values
print(np.sum(sentence_NaN), '\n') #see how many misisng values there are: 9053

print(pd.crosstab(sentence_NaN, charge), '\n') #missing cases are shown as a value of 4
sentence = sentence.mask(charge == 4, 0) #replaces missing values with a 0
sentence = sentence.mask(charge == 9, np.nan) #replaces the unknown 9 value as missing

sentence_NaN = sentence.isnull() #new dummy
print(pd.crosstab(sentence_NaN, charge), '\n') # now only 274 missing values

trial['ImposedSentenceAllChargeInContactEvent'] = sentence #replace the data with the clean version
```
